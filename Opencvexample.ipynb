```python
def resizeFrame(frame,scale=0.75):
    print(f'original frame size is{frame.shape[1],frame.shape[0]} ')
    width=int(frame.shape[1]*scale)
    height=int(frame.shape[0]*scale)
   
    dim=(width,height)
    return cv.resize(frame,dim,interpolation=cv.INTER_AREA)
```

```python
#Morphological transformation are simple operationbased on the image shape.Normally perfomed on binary images.
#A kernel tells how to change the value of given pixel by combining it with different amount of neighbouring pixel

import numpy as np
img = cv.imread(path,cv.IMREAD_GRAYSCALE)
_,mask = cv.threshold(img,220,255,cv.THRESH_BINARY_INV)
kernal=np.ones((5,5),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=2)
erosion=cv.erode(mask,kernal,iterations=2)
opening=cv.morphologyEx(mask,cv.MORPH_OPEN,kernal)
closing=cv.morphologyEx(mask,cv.MORPH_CLOSE,kernal)
gradient=cv.morphologyEx(mask,cv.MORPH_GRADIENT,kernal)
tophat=cv.morphologyEx(mask,cv.MORPH_BLACKHAT,kernal)
titles=["images","mask","dilation","erosion","opening","closing","gradient","blackhat"]
images=[img,mask,dilation,erosion,opening,closing,gradient,tophat]

for i in range(8):
    plt.figure(figsize=(40,40))
    plt.subplot(1,8,i+1)
    plt.imshow(images[i],"gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.show()
```

```python
#Load the two image
# Find the gaussian pyramid for mango and orange 
# from gaussian pyramid find the laplacian pyramid
# now join the left of mango and right of orange in each level of laplacian pyramid
#finally from this joint image pyramids , reconstruct the original image
```

```python
#Load the two image
mango=cv.imread(mpic)
orange=cv.imread(opic)
print(mango.shape)
print(orange.shape)


mangocopy=mango.copy()
orangecopy=orange.copy()
gp_mango=[mangocopy]
gp_orange=[orangecopy]

mango_orange=np.hstack((mango[:,:280],orange[:,280:]))
```

```python
#generate gaussian pyramid for mango
for i in range(6):
    mangocopy=cv.pyrDown(mangocopy)
    gp_mango.append(mangocopy)
    
#generate gaussian pyramid for orange
for i in range(6):
    orangecopy=cv.pyrDown(orangecopy)
    gp_orange.append(orangecopy)

    
#generate laplacian pyramid for mango
mangocopy=gp_mango[5]
lp_mango=[mangocopy]
for i in range(5,0,-1):
    guassian_expanded=cv.pyrUp(gp_mango[i])
    laplacian=cv.subtract(gp_mango[i-1],guassian_expanded)
    lp_mango.append(laplacian)

    
#generate laplacian pyramid for mango


    

```

```python
orangecopy=gp_orange[5]
lp_orange=[orangecopy]
for i in range(5,0,-1):
    guassian_expanded=cv.pyrUp(gp_orange[i])
    laplacian=cv.subtract(gp_orange[i-1],guassian_expanded)
    lp_orange.append(laplacian)
    
#add left and right halves of images in each level
mango_orange_py=[]
n=0
for mango_lp,orange_lp in zip(lp_mango,lp_orange):
    n+=1
    cols,rows,ch=mango_lp.shape
    lp=np.hstack((mango_lp[:,0:int(cols/2)],orange_lp[:,int(cols/2):]))
    mango_orange_py.append(lp)

#reconstruct original image
mango_orange_recon=mango_orange_py[0]
for i in range(1,6):
    mango_orange_recon=cv.pyrUp(mango_orange_recon)
    mango_orange_recon=cv.add(mango_orange_py[i],mango_orange_recon)

    
cv.imshow('mango',mango)
cv.imshow('orange',orange)
cv.imshow('mango_orange',mango_orange)
cv.imshow('mango_orange_reconstruct',mango_orange_recon)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
vpath="C:\\Users\\Aayush Shukla\\OneDrive\\Desktop\\walking.mp4"
cap = cv.VideoCapture(vpath)
_,frame1=cap.read()
_,frame2=cap.read()

while cap.isOpened():
    diff=cv.absdiff(frame1,frame2)
    gray=cv.cvtColor(diff,cv.COLOR_BGR2GRAY)
    blur=cv.GaussianBlur(gray,(5,5),0)
    _,thresh=cv.threshold(blur,20,255,cv.THRESH_BINARY)
    dilated=cv.dilate(thresh,None,iterations=3)
    contours,_=cv.findContours(dilated,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for controur in contours:
        (x,y,w,h)= cv.boundingRect(controur)
        if cv.contourArea(controur)<900:
            continue 
        cv.rectangle(frame1,(x,y),(x+w,y+h),(0,255,0),2)
        cv.putText(frame1,"Status {}".format('Movement'),(20,20),cv.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
    #cv.drawContours(frame1,contours,-1,(0,255,0),2)
    cv.imshow("frame",frame1)
    frame1=frame2
    ret,frame2 = cap.read()
 
    if cv.waitKey(20) & 0xFF==ord('d'):
        break

cv.destroyAllWindows()
cap.release()

```

```python
simg=cv.imread(ipath)
sgray=cv.cvtColor(simg,cv.COLOR_BGR2GRAY)
_,thresh=cv.threshold(sgray,240,255,cv.THRESH_BINARY)
contours ,_ =cv.findContours(thresh,cv.RETR_TREE,cv.CHAIN_APPROX_NONE)
for contour in contours:
    approx=cv.approxPolyDP(contour,0.01*cv.arcLength(contour,True),True)
    print(approx)
    cv.drawContours(simg,approx,-1,(0,255,0),5)
    x=approx.ravel()[0]
    y=approx.ravel()[1]
    if len(approx)==3:
        cv.putText(simg,'Triangle',(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,180,20),1)
    elif len(approx)==5:
        cv.putText(simg,'Pentagon',(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,180,20),1)
    elif len(approx)==4:
        cv.putText(simg,'Square',(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,180,20),1)
    elif len(approx)==6:
        cv.putText(simg,'Hexagon',(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,180,20),1)
    else:
        cv.putText(simg,'Circle',(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,180,20),1)
        
    
    
cv.imshow('original',simg)
cv.imshow('thresh',thresh)
cv.waitKey(0)
cv.destroyAllWindows()
```
